```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
taxi_data = pd.read_csv('Main_Sample.csv')

# Convert pickup and dropoff times to datetime objects
taxi_data['tpep_pickup_datetime'] = pd.to_datetime(taxi_data['tpep_pickup_datetime'])
taxi_data['tpep_dropoff_datetime'] = pd.to_datetime(taxi_data['tpep_dropoff_datetime'])

# Extract the day of the week from the pickup time
taxi_data['pickup_day'] = taxi_data['tpep_pickup_datetime'].dt.day_name()

# Create a list of weekdays and weekends
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekends = ['Saturday', 'Sunday']

# Calculate the average revenue for weekdays and weekends
weekday_revenue = taxi_data[taxi_data['pickup_day'].isin(weekdays)]['total_amount'].mean()
weekend_revenue = taxi_data[taxi_data['pickup_day'].isin(weekends)]['total_amount'].mean()

# Calculate the mean value of both bars separately
weekday_mean = taxi_data[taxi_data['pickup_day'].isin(weekdays)]['total_amount'].mean()
weekend_mean = taxi_data[taxi_data['pickup_day'].isin(weekends)]['total_amount'].mean()

# Create a bar chart
plt.bar(['Weekdays', 'Weekends'], [weekday_revenue, weekend_revenue])

# Add labels, title, and mean values to the bar chart
plt.xlabel('Day of the Week')
plt.ylabel('Average dollars Revenue')
plt.title('Average Revenue by Weekdays and Weekends\nWeekday Mean Value: ${:.2f}\nWeekend Mean Value: ${:.2f}'.format(weekday_mean, weekend_mean))

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

# Load the CSV file into a pandas dataframe
taxi_data = pd.read_csv('Main_Sample.csv')

# Convert pickup and dropoff times to datetime objects
taxi_data['tpep_pickup_datetime'] = pd.to_datetime(taxi_data['tpep_pickup_datetime'])
taxi_data['tpep_dropoff_datetime'] = pd.to_datetime(taxi_data['tpep_dropoff_datetime'])

# Extract the day of the week from the pickup time
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
taxi_data['pickup_day'] = pd.Categorical(taxi_data['tpep_pickup_datetime'].dt.day_name(), categories=days_of_week, ordered=True)

# Create a dictionary to store the average revenue for each day of the week
daily_revenue = {}
for day in days_of_week:
    # Calculate the average revenue for the day
    daily_revenue[day] = taxi_data[taxi_data['pickup_day'] == day]['total_amount'].mean()

# Create a bar chart for each day of the week
fig, ax = plt.subplots()
ax.bar(daily_revenue.keys(), daily_revenue.values(), width=0.6)

# Format y-axis labels with two decimal places
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

# Add labels and title to the bar chart
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Average Revenue')
ax.set_title('Average dollars Revenue by Day of the Week')

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the Main Sample data
data = pd.read_csv('Main_Sample.csv')

# Split the data into training and testing sets
train_data = data.iloc[:int(0.8*len(data))]
test_data = data.iloc[int(0.8*len(data)):]

# Create the predictor variables and target variable
X_train = train_data[['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount']]
y_train = train_data['total_amount']
X_test = test_data[['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount']]
y_test = test_data['total_amount']

# Create the linear regression model and fit it to the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate the RMSE and R2 score
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the RMSE and R2 score
print("RMSE:", rmse)
print("R^2 score:", r2)

# Predict the total amount for the trips in the New Sample file
new_data = pd.read_csv('New_Sample.csv')
X_new = new_data[['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount']]
y_new_pred = model.predict(X_new)
print("Predicted Total Amount:")
print(y_new_pred)

# Create a scatter plot of actual vs predicted total amounts
plt.scatter(new_data.index, new_data['tip_amount'], color='blue', label='Actual')
plt.scatter(new_data.index, y_new_pred, color='red', label='Predicted')
plt.xlabel('Trip')
plt.ylabel('Total Amount')
plt.title('Actual vs Predicted Total Amounts')
plt.legend()
plt.show()

```
